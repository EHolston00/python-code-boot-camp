```python
def playGame(userChoice, cpuChoice) :
    if ((cpuChoice == 'rock' and userChoice == 'paper') or 
        (cpuChoice == 'paper' and userChoice == 'scissors') or
        (cpuChoice == 'scissors' and userChoice == 'rock')) :
        print(f"You chose: {userChoice}\nCPU chose: {cpuChoice}\n")
        print("You win!\n")
        return 'user'
        
    elif ((cpuChoice == 'paper' and userChoice == 'rock') or 
          (cpuChoice == 'scissors' and userChoice == 'paper') or
          (cpuChoice == 'rock' and userChoice == 'scissors')) :
        print(f"You chose: {userChoice}\nCPU chose: {cpuChoice}\n")  
        print("You lose.\n")
        return 'cpu'  
              
    else :
        print(f"You chose: {userChoice}\nCPU chose: {cpuChoice}\n")
        print("It's a draw.\n")
        return 'draw'
          
def isInvalidContinueInput(userInput) :
    return userSelection != 'y' and userSelection != 'Y' and userSelection != 'n' and userSelection != 'N'
          
def updateScore(dict, winner) :
    if winner == 'user' :
        scores['user'].append(1)
        scores['cpu'].append(0)
    elif winner == 'cpu' :
        scores['user'].append(0)
        scores['cpu'].append(1)
    else :
        scores['user'].append(0)
        scores['cpu'].append(0)
          
          
def printTotalScore(dict) :
    userPoints = 0
    cpuPoints = 0
    
    for userScore in dict['user'] :
        userPoints += userScore
          
    for cpuScore in dict['cpu'] :
        cpuPoints += cpuScore
          
    print(f"Current score\n\nUser: {userPoints}\nCPU: {cpuPoints}\n")
              

def showRoundHistory(dict) :
    totalUserScore = 0
    totalCPUScore = 0
    
    for x in range(0, len(dict['user'])) :
        userRoundValue = dict['user'][x]
        cpuRoundValue = dict['cpu'][x]
        totalUserScore += userRoundValue
        totalCPUScore += cpuRoundValue
        
        print(f"Round {x+1} Results: ", end='\t')
        print(f"User: {totalUserScore}\tCPU: {totalCPUScore}", end='\t')
        
        if(userRoundValue > cpuRoundValue) :
          print("You won!\n")
        elif (cpuRoundValue > userRoundValue) :
          print("You lost.\n")
        else :
          print("It was a draw\n")
        
          
```

```python
import random as rand

options = ('rock', 'paper', 'scissors')
scores = {'user': [], 'cpu': []}

userSelection = 'y'
while True:
    computerSelection = options[rand.randint(0,2)]
    userSelection = input("rock, paper, or scissors: ")

    if userSelection not in options :
        print("***Invalid selection. Please choose between the following options (rock/paper/scissors).\n")
        continue
    else :
        # play the game, update score and print the current total score
        winner = playGame(userSelection, computerSelection)
        updateScore(scores, winner)
        printTotalScore(scores)
        
        # if the user is inputs anything other than y/n/Y/N, prompt them to enter a valid option
        while isInvalidContinueInput(userSelection) :
            userSelection = input("Would you like to play again (y/n)? ")
            if isInvalidContinueInput(userSelection) :
                print("***Invalid selection. Please choose between the following options (y/n).\n")
                continue
                
        if(userSelection == 'y' or userSelection == 'Y') :
            continue
        break;
                

showRoundHistory(scores)
print("Thank you for playing!")
```
